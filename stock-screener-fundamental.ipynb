```python
def gordonGrowthModelReturn(payoutRatio, PE, returnOnEquity):
    return payoutRatio / PE + returnOnEquity*(1 - payoutRatio)

def dividend10YearReturn(payoutRatio, PE, returnOnEquity):
    if PE <= 0:
        return -1
    
    terminalPE = min(PE, 12)
    b = 1 - payoutRatio
    p = payoutRatio
    g = returnOnEquity *  b
    def func(r):
        return (1+g)**1 * p / (1 + r)**1 + \
        (1+g)**2 * p / (1 + r)**2 + \
        (1+g)**3 * p / (1 + r)**3 + \
        (1+g)**4 * p / (1 + r)**4 + \
        (1+g)**5 * p / (1 + r)**5 + \
        (1+g)**6 * p / (1 + r)**6 + \
        (1+g)**7 * p / (1 + r)**7 + \
        (1+g)**8 * p / (1 + r)**8 + \
        (1+g)**9 * p / (1 + r)**9 + \
        (1+g)**10 * p / (1 + r)**10 + \
        (1+g)**11 * terminalPE / (1 + r)**11 \
        - PE
           
    root = fsolve(func, [0])
    return root[0]

def rdpModelReturn(payoutRatio, PE, returnOnEquity):
    return (1 + returnOnEquity*(1-payoutRatio))*(1+payoutRatio/PE)-1
```

```python
stock_data = {}

f = open('stock-screen.csv', 'w')
f.write('ticker|returnOnEquity|returnOnAssets|payoutRatio|PE|debtToEquity|growthRate|ggmReturn|div10YReturn|rdpReturn\n')
for ticker in tickers:
    print(ticker)
    url = 'https://query1.finance.yahoo.com/v10/finance/quoteSummary/' + ticker + '?modules=summaryDetail,financialData'
    r = requests.get(url)
    if r.json()['quoteSummary']['result'] is None:
        continue
    
    data = r.json()['quoteSummary']['result'][0]
    
    if 'financialData' not in data or  'summaryDetail' not in data:
        continue
    
    if 'raw' in data['financialData']['returnOnEquity']:
        returnOnEquity = data['financialData']['returnOnEquity']['raw']
    else:
        returnOnEquity = 0
    
    if 'raw' in data['financialData']['returnOnAssets']:
        returnOnAssets = data['financialData']['returnOnAssets']['raw']
    else:
        returnOnAssets = 0  
        
    if 'raw' in data['summaryDetail']['payoutRatio']:
        payoutRatio = data['summaryDetail']['payoutRatio']['raw']
    else:
        payoutRatio = 0
        
    if 'trailingPE' in data['summaryDetail']:
        PE = data['summaryDetail']['trailingPE']['raw']
    elif 'raw' in data['summaryDetail']['forwardPE']:
        PE = data['summaryDetail']['forwardPE']['raw'] 
    else:
        PE = 99999
        
    if isinstance(PE, str):
        PE = 99999
    
    if 'raw' in data['financialData']['debtToEquity']:
        debtToEquity = data['financialData']['debtToEquity']['raw']
    else:
        debtToEquity = 0
    
    growthRate = returnOnEquity * (1 - payoutRatio)
    
    ggmReturn = gordonGrowthModelReturn(payoutRatio, PE, returnOnEquity)
    div10YReturn = dividend10YearReturn(payoutRatio, PE, returnOnEquity)
    rdpReturn = rdpModelReturn(payoutRatio, PE, returnOnEquity)
    
    stock_data[ticker] = {
        'returnOnEquity': float('{:.4f}'.format(returnOnEquity)),
        'returnOnAssets': float('{:.4f}'.format(returnOnAssets)),
        'payoutRatio': float('{:.4f}'.format(payoutRatio)),
        'PE': float('{:.4f}'.format(PE)),
        'debtToEquity': float('{:.4f}'.format(debtToEquity)),
        'growthRate': float('{:.4f}'.format(growthRate)),
        'ggmReturn': float('{:.4f}'.format(ggmReturn)),
        'div10YReturn': float('{:.4f}'.format(div10YReturn)),
        'rdpReturn': float('{:.4f}'.format(rdpReturn)),
    }
    
    f.write(ticker + '|' + 
            '{:.4f}'.format(returnOnEquity) + '|' +
            '{:.4f}'.format(returnOnAssets) + '|' +
            '{:.4f}'.format(payoutRatio) + '|' +
            '{:.4f}'.format(PE) + '|' +
            '{:.4f}'.format(debtToEquity) + '|' +
            '{:.4f}'.format(growthRate) + '|' +
            '{:.4f}'.format(div10YReturn) + '|' +
            '{:.4f}'.format(rdpReturn) + '\n'
           )
f.close()
```
